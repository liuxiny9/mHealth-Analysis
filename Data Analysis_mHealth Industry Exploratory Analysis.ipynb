```python
#convert currency into USD for funding amount comparsion
#column contains currency information 'Last Funding Amount','Total Funding Amount','Last Equity Funding Amount','Total Equity Funding Amount'
#identify unique currency used in the dataset 

import re
def extract_currency_symbol(amount):
    match = re.search(r'[^\d,.]+', amount)
    return match.group(0).strip() if match else 'USD'  # Default to USD if no symbol is found

currency_columns = ['Last Funding Amount', 'Total Funding Amount', 'Last Equity Funding Amount', 'Total Equity Funding Amount']
currency_symbols_1 = set()

for columns in currency_columns:
    currency_symbols_1.update(df_1[columns].dropna().apply(extract_currency_symbol))

currency_symbols_1
```

```python
#identify exchange rate for identified currency code
exchange_rates = {
    'USD': 1.0,
    '$': 1.0,
    'CHF': 1.17,  # Swiss Franc to USD
    '₣': 1.17,
    'GBP': 1.28,  # British Pound to USD
    '£': 1.28,
    'KRW': 0.00073,  # South Korean Won to USD
    '₩': 0.00073,
    'EUR': 1.10,  # Euro to USD
    '€': 1.10,
    'INR': 0.012,  # Indian Rupee to USD
    '₹': 0.012,
    '—': 1.0  # Assume missing or unknown values are already in USD
}

#define function to convert other currency to USD 
#logic: 
#Step 1
#add one additional column to the orignial dataset, stating the original currency.
#for example, TOM Medications should have a column named currency, the value will be 'CHF'

df_1['Total_Funding_Amount_Currency']= df_1 ['Total Funding Amount'].apply(extract_currency_symbol)

#Step 2
#Create a function to apply, logic is the following: 
#If the currency is listed in the exchange rate table, then use the exchange rate mapping and apply the original amount * the exchange rate, 
#if the currency is not found in the exchange rate table, then return the original amount 

# Define the conversion function
def convert_to_usd(amount, currency, exchange_rates):
    try:
        # Remove non-numeric characters (except for decimal point) and convert to float
        value = float(re.sub(r'[^\d.]', '', str(amount)))
    except ValueError:
        return amount  # If conversion fails, return the original amount
    if currency in exchange_rates:
        return value * exchange_rates[currency] 
    else:
        return amount  # Return the original amount if currency is not in exchange rates

# Apply the conversion function
df_1['Total Funding Amount (USD)'] = df_1.apply(
    lambda row: convert_to_usd(row['Total Funding Amount'], row['Total_Funding_Amount_Currency'], exchange_rates),
    axis=1
)


df_1
```

```python
#Apply the same currency conversion to dataset 2

currency_symbols_2 = set()

for columns in currency_columns:
    currency_symbols_2.update(df_2[columns].dropna().apply(extract_currency_symbol))

currency_symbols_2
```

```python
df_2['Total_Funding_Amount_Currency']= df_2 ['Total Funding Amount'].apply(extract_currency_symbol)

# Apply the conversion function
df_2['Total Funding Amount (USD)'] = df_2.apply(
    lambda row: convert_to_usd(row['Total Funding Amount'], row['Total_Funding_Amount_Currency'], exchange_rates),
    axis=1
)
df_2['Total Funding Amount (USD)'] = df_2['Total Funding Amount (USD)'].apply(lambda x:format(x, ' .0f'))
df_2
```

```python
#Re-Rank Companies base on Total Funding Amount (USD) Desc and rename dataset

df_1_sorted = df_1.sort_values(by = 'Total Funding Amount (USD)', ascending = False) 
df_1_sorted['Total Funding Amount (USD)'] = pd.to_numeric(df_1_sorted['Total Funding Amount (USD)'], errors='coerce')
```

```python
df_2_sorted = df_2.sort_values(by = 'Total Funding Amount (USD)', ascending = False) 
df_2_sorted['Total Funding Amount (USD)'] = pd.to_numeric(df_2_sorted['Total Funding Amount (USD)'], errors='coerce')
```

## Generate General Statistics base on Total Fundings Information 

```python
# Generate General Statistics base on total fundings information 

statistics_mHealth = df_1_sorted ['Total Funding Amount (USD)'].describe()
#statistics_mHealth_formatted = statistics_mHealth.apply(lambda x: f"{x:,.2f}")

statistics_healthcare = df_2_sorted ['Total Funding Amount (USD)'].describe()
#statistics_healthcare_formatted = statistics_healthcare.apply(lambda x: f"{x:,.2f}")

combined_statistics = pd.DataFrame({
    'Statistic': ['Count', 'Mean', 'Standard Deviation', 'Min', '25%', '50% (Median)', '75%', 'Max'],
    'mHealth': statistics_mHealth.values,
    'healthcare': statistics_healthcare.values
})

combined_statistics['Comparison'] = combined_statistics['healthcare'] / combined_statistics['mHealth']
combined_statistics['mHealth'] = combined_statistics['mHealth'].apply(lambda x: f"{int(x):,}" if pd.notnull(x) else 'NaN')
combined_statistics['healthcare'] = combined_statistics['healthcare'].apply(lambda x: f"{int(x):,}" if pd.notnull(x) else 'NaN')
combined_statistics['Comparison'] = combined_statistics['Comparison'].apply(lambda x: f"{int(x):,}" if pd.notnull(x) else 'NaN')

combined_statistics
```

```python
#Draw historgram to visualize the distribution of total funding value 
from matplotlib.ticker import FuncFormatter
def millions(x, pos):
    'The two args are the value and tick position'
    if x >= 1e9:
        return f'${x * 1e-9:.1f}B'
    else:
        return f'${x * 1e-6:.0f}M'

formatter = FuncFormatter(millions)

plt.figure(figsize = (10,5))
df_1_sorted['Total Funding Amount (USD)'].hist(bins = 20, alpha = 0.5, color = 'blue', label = 'mHealth')
plt.title('Histogram of Total Funding Distribution (USD) for mHealth Company')
plt.xlabel('Total Funding Amount (USD)')
plt.ylabel('Frequency')
plt.legend()
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize = (10,5))
df_2_sorted['Total Funding Amount (USD)'].hist(bins = 20, alpha = 0.5, color = 'blue', label = 'healthcare')
plt.title('Histogram of Total Funding Distribution (USD) for healthcare Company')
plt.xlabel('Total Funding Amount (USD)')
plt.ylabel('Frequency')
plt.legend()
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Is there any mHealth companies made to top 50 of overall healthcare company?

common_indices = df_1_sorted.index.intersection(df_2_sorted.index)
common_indices

```
